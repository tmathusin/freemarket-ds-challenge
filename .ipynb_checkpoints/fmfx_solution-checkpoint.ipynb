```python
%load_ext autoreload
%autoreload 2
    
import math
import os, sys
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import networkx as nx
import numpy as np
sys.path.append(os.path.abspath('Code Test'))
from network import build_flow_pairs, participant_metrics, top_participants, build_counterparty_metrics, classify_quadrants, build_client_network
from cleaning import standardise_counterparty_names, aggregate_flows
from viz import bar_top_series, plot_network

data_path = 'data/Test for Data Science role.xlsx'
```

```python
# Load all tables
# Returns a dict of dataframes with keys: client, accounts, deposits, withdrawals, transfers
sheets = ['Client', 'Accounts', 'Deposits', 'Withdrawals', 'Transfers']
dfs = {s.lower(): pd.read_excel(data_path, sheet_name=s, header=1) for s in sheets}

# Normalise column names
for k, df in dfs.items():
    # Drop columns that are entirely NaN
    df = df.dropna(axis=1, how='all')
    # df.columns = (df.columns.str.strip().str.lower())
    df.columns = (df.columns
                  .str.replace('(?<=[a-z])(?=[A-Z])', '_', regex=True)
                  .str.lower()
                 )
    dfs[k] = df
```

```python
# List all columns by table

for k,v in dfs.items():
    print(k, v.shape, list(v.columns))
```

```python
# assign dataframe name to each table
clients_df = dfs['client']
accounts_df = dfs['accounts']
deposits_df = dfs['deposits'].iloc[:, :-1]
withdrawals_df = dfs['withdrawals']
transfers_df = dfs['transfers']
```

## 2) Outside the Network — Fuzzy standardisation

```python
# Deposits (Remitters)
deposits_df.columns
```

```python
# Deposits (Remitters)
deposits_std = standardise_counterparty_names(deposits_df, col="deposit_remitter_name", threshold=0.9)
```

```python
deposits_std.head()
```

```python
remitter_flows = aggregate_flows(deposits_std, entity_col="deposit_remitter_name_standardised", amount_col="normalised_amount")
remitter_flows.head(10)
```

```python
# Withdrawals (Beneficiaries)
withdrawals_df.columns
```

```python
# Withdrawals (Beneficiaries)
withdrawals_std = standardise_counterparty_names(withdrawals_df, col='beneficiary_name') # withdrawals table with standardise_counterparty_names
withdrawals_std.head()
```

```python
beneficiary_flows = aggregate_flows(withdrawals_std, entity_col="beneficiary_name_standardised",amount_col="normalised_amount")
beneficiary_flows.head(10)
```

```python
bar_top_series(remitter_flows, 'entity', 'amount', 15, title='Top remitters by consolidated flow')
plt.show()
bar_top_series(beneficiary_flows, 'entity', 'amount', 15, title='Top beneficiaries by consolidated flow')
plt.show()
```

```python
# --- Top remitters (Deposits) & beneficiaries (Withdrawals) by VOLUME ---

TOP_N = 20

# # Guard for missing columns
# has_remitter = "deposit_remitter_name_standardised" in deposits_std.columns
# has_beneficiary = "beneficiary_name_standardised" in withdrawals_std.columns

top_remitters_by_volume = (
    deposits_std
    .assign(_name=lambda d: d["deposit_remitter_name_standardised"].fillna("UNKNOWN_REMITTER"))
    .groupby("_name", as_index=False)
    .agg(tx_count=("normalised_amount", "size"))
    .sort_values("tx_count", ascending=False)
    .head(TOP_N)
    .rename(columns={"_name": "remitter"})
) 

top_beneficiaries_by_volume = (
    withdrawals_std
    .assign(_name=lambda d: d["beneficiary_name_standardised"].fillna("UNKNOWN_BENEFICIARY"))
    .groupby("_name", as_index=False)
    .agg(tx_count=("normalised_amount", "size"))
    .sort_values("tx_count", ascending=False)
    .head(TOP_N)
    .rename(columns={"_name": "beneficiary"})
) 

# display(top_remitters_by_volume)
# display(top_beneficiaries_by_volume)
```

```python
if not top_remitters_by_volume.empty:
    plt.figure()
    plt.barh(top_remitters_by_volume["remitter"].head(15)[::-1],
             top_remitters_by_volume["tx_count"].head(15)[::-1])
    plt.title("Top Remitters by Volume")
    plt.xlabel("Transactions")
    plt.tight_layout()

if not top_beneficiaries_by_volume.empty:
    plt.figure()
    plt.barh(top_beneficiaries_by_volume["beneficiary"].head(15)[::-1],
             top_beneficiaries_by_volume["tx_count"].head(15)[::-1])
    plt.title("Top Beneficiaries by Volume")
    plt.xlabel("Transactions")
    plt.tight_layout()
```

## 3. Build counterparty metrics (Remitters & Beneficiaries)

```python
# Beneficiaries: withdrawals_std
ben_metrics = build_counterparty_metrics(
    withdrawals_std,
    entity_col="beneficiary_name_standardised",
    amount_col="normalised_amount",
    role="beneficiary",
)

# Remitters: deposits_std
rem_metrics = build_counterparty_metrics(
    deposits_std,
    entity_col="deposit_remitter_name_standardised",
    amount_col="normalised_amount",
    role="remitter",
)

# Combined view
metrics_all = pd.concat([rem_metrics, ben_metrics], ignore_index=True)

# Save
metrics_all.to_csv("output/counterparty_metrics.csv")
```

## 4. Classify into quadrants

```python
rem_quads = classify_quadrants(rem_metrics)
ben_quads = classify_quadrants(ben_metrics)
all_quads = classify_quadrants(metrics_all)

# Save
all_quads.to_csv("output/counterparty_quadrants.csv", index=False)
```
